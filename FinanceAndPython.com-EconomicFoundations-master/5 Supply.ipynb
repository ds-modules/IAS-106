```python
import matplotlib.pyplot as plt
prices = []
hamburgers = []

for price in range(0,11):
    prices += [price]
    hamburgers += [price]
plt.plot(hamburgers,prices)
plt.xlabel("Quantity Supplied")
plt.ylabel("Price")
plt.show()
```

```python
prices = []
hamburgers = []

for price in range(0,11):
    prices += [price]
    hamburgers += [price]
plt.plot(hamburgers,prices)




prices = []
hamburgers = []

for price in range(0,11):
    prices += [price]
    hamburgers += [price*2]
plt.plot(hamburgers,prices)




prices = []
hamburgers = []

for price in range(0,11):
    prices += [price]
    hamburgers += [(price)+(2*price)]
plt.plot(hamburgers,prices)

plt.xlabel("Quantity Supplied")
plt.ylabel("Price")
plt.title("Combined Supply of Hamburgers")
plt.legend(['Producer 1','Producer 2','Combined'])
plt.show()
```
