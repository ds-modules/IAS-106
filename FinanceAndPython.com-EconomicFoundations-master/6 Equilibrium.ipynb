```python
import sympy
p = sympy.Symbol("p")

demandEquation = 10-p
supplyEquation = p

prices = []
demandQ = []
supplyQ = []

for price in range(0,11):
    demandQ += [demandEquation.subs(p,price)]
    supplyQ += [supplyEquation.subs(p,price)]
    prices += [price]
print(prices)
print(demandQ)
print(supplyQ)
```

```python
EquilibriumP = sympy.solve(demandEquation-supplyEquation)
print(EquilibriumP)
EquilibriumP = EquilibriumP[0]
print(EquilibriumP)
```

```python
EquilibriumQ = demandEquation.subs(p,EquilibriumP)
print(EquilibriumQ)
```

```python
import matplotlib.pyplot as plt
plt.plot(demandQ,prices)
plt.plot(supplyQ,prices)
plt.plot(EquilibriumQ,EquilibriumP, 'ro')

plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(demandQ,prices)
plt.plot(supplyQ,prices)
unitsDemanded = demandEquation.subs(p,8)
unitsSupplied = supplyEquation.subs(p,8)
plt.plot(unitsDemanded,8, 'ro')
plt.plot(unitsSupplied,8, 'ro')
plt.plot([unitsDemanded,unitsSupplied],[8,8],":")

plt.xlabel("Supply and Demand")
plt.ylabel("Price")
plt.show()

surplus = unitsSupplied-unitsDemanded

print("The surplus is equal to "+str(surplus)+" units.")
```
