```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# reading in the data
df = pd.read_csv('econ.csv')
# converting column types
df['ATC'] = df['ATC'].str.replace(',', '')
df['ATC'] = df['ATC'].astype('int64')
df['AFC'] = df['AFC'].str.replace(',', '')
df['AFC'] = df['AFC'].astype('int64')
df['AVC'] = df['AVC'].str.replace(',', '')
df['AVC'] = df['AVC'].astype('int64')
df.head()
```

```python
# plotting TC FC VC
plt.plot(df['Q'], df['TC'], color='blue')
plt.plot(df['Q'], df['FC'], color='red')
plt.plot(df['Q'], df['VC'], color='orange')
plt.title('TC FC VC')
plt.xlabel('Quantity of Production')
plt.ylabel('$$$')
plt.legend()
plt.show()
```

```python
# plotting ATC AFC AVC
plt.plot(df['Q'], df['ATC'], color='blue')
plt.plot(df['Q'], df['AFC'], color='red')
plt.plot(df['Q'], df['AVC'], color='orange')
plt.title('ATC AFC AVC')
plt.xlabel('Quantity of Production')
plt.ylabel('$/unitoutput')
plt.legend()
plt.show()
```
